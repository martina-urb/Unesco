```python

def fetch_image_url(site_name):
    """Cerca l'URL dell'immagine principale di un sito su Wikipedia."""
    search_url = f"https://en.wikipedia.org/wiki/{site_name.replace(' ', '_')}"
    try:
        response = requests.get(search_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Cerca l'immagine nell'infobox
            image_tag = soup.find('table', {'class': 'infobox'}).find('img')
            if image_tag:
                return f"https:{image_tag['src']}"  # Costruisci l'URL assoluto
        return None  # Nessuna immagine trovata
    except Exception as e:
        print(f"Errore con {site_name}: {e}")
        return None

# Esempio di siti
heritage_sites = ["Taj Mahal", "Great Wall of China", "Machu Picchu"]

# Scarica immagini per ogni sito
for site in heritage_sites:
    fetch_image(site)
```
